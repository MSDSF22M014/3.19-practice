```python
#what if we have 2 or more temperatures on the same date of a particular city

import pandas as pd
df = pd.read_csv('datasets/pivot_weather2.csv')
df
```

```python
#here 'pivot fails'
#df.pivot(index = 'city', columns = 'date')
#df.pivot(index = 'date', columns = 'city')
```

```python
#pivot table comes to rescue
#by default the aggfunc is mean
df.pivot_table(index = 'city', columns = 'date')
```

```python

df.pivot_table(index = 'date', columns = 'city')
```

```python
#we can change the aggfunc
df.pivot_table(index = 'city', columns = 'date', aggfunc = 'sum')
```

```python
# mean of all the column values can be added by 'margins = True'. add an additional column and row 'all'
df.pivot_table(index = 'city', columns = 'date', margins = True)
```

```python
#using melt. data ko long karta hay wide kay bajaye
df = pd.read_csv('datasets/weather.csv')
df
```

```python
df1 = pd.melt(df, id_vars =['day'])
df1
```

```python
# df1 = pd.pivot_table(df, index = 'day')
# df1
```

```python
#we can give meaningful names
df1 = pd.melt(df, id_vars =['day'], var_name='city', value_name='temperature')
df1
```

```python
# we can choose data of our choice using 'value_vars'
df2 = pd.melt(df, id_vars = ['day'], value_vars =['karachi'], var_name='city', value_name='temperature')
df2
```

```python
#old fashioned way
df1[df1['city'] == 'karachi']
```

```python
# using an agg func on an entire df
df1['temperature'].agg('mean')
```

```python
#crosstab can be used, normally used for comparing categorical cols
# Reading data from 'datasets/sample.csv' file
import numpy as np
import pandas as pd
df = pd.read_csv('datasets/sample1.csv')
df.head()
```

```python
pd.crosstab(index=df.city, columns=df.gender)
#ye bata deta hay kay kis city mein kitny male female hain.count bataye ga
```

```python
#cannot be used like this
#df.crosstab(index='city', columns='gender')
```

```python
pd.crosstab(index=df.gender, columns=df.city)
```

```python
#agr percentage chahiye ho to ye karo
pd.crosstab(index=df.city, columns=df.gender, normalize=True)
```

```python
# agar average age chahiye ho different city mein male female ki to...
pd.crosstab(index=df.city, columns=df.gender, values=df.age, aggfunc=np.mean)
```
